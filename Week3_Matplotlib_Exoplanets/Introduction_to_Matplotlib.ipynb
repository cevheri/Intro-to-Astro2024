## Basic 1D Plots

### Line Plots

Suppose we want to plot the function $y = x^2$. To do that, we need data points, i.e, coordinate points. These are made using numpy.

```python
x_values = np.linspace(0,10,5)
x_values# see as discussed in the previous lecture, it creates 5 numbers between 0 to 10
```

```python
y_values = x_values**2
print(y_values)
```

```python
plt.plot(x_values,y_values)
plt.show()#to show the plot. In jupyter notebook we don't need this actually but it's a good practice
```

See because there are only 5 points and they are connected by lines, They don't look too good.

**So, Let's increase the number of points**

```python
x_values = np.linspace(0,10,100)
y_values = x_values**2
plt.plot(x_values,y_values)
plt.show()## Smooth curve...
```

To change the plotting style, use the syntax

```python
plt.plot(x_values,y_values, color="Red", ls="--")#ls means line-style
plt.show()## Smooth curve...
```

supported values are **'-'**, **'--'**, **'-.'**, **':'**, **'None'**, **' '**, **''**, **'solid'**, **'dashed'**, **'dashdot'**, **'dotted'**

If you want to see the points along with line use,

```python
x = np.linspace(-2,2,10)
y = x**2
plt.plot(x,y, 'o--', color='red', lw=2, ms=10) #lw is line width, ms is marker size
```

Equivalently, you can use the syntax

```python
plt.plot(x,y, linestyle='--', marker='o', color='red', linewidth=2, markersize=10) 
```

The above syntax can be useful as a way of self-documenting your code. As you continue to progress as a programmer and a researcher, you will learn that documenting code is very important.

To change the figure size

```python
plt.figure(figsize=(8,3))
plt.plot(x,y)
```

When we have more than one plot, we need to label things. We can also add title and axis label.

```python
x = np.linspace(-2,2,100)
y = 4*np.sin(x)
y1 = 4*np.cos(x)
```

```python
plt.figure(figsize=(8,3))
plt.plot(x,y1,ls="--", label=r'$\sin(x)$') # The $\sin(x)$ is Latex syntax. You will learn more about this later in the course.
plt.plot(x,y, label=r'$\cos(x)$')
plt.xlabel('x-axis', fontsize=16)
plt.ylabel('y-axis', fontsize=16)
plt.title('Plots of Different Functions')
plt.legend(loc='lower right', fontsize=12, ncol=1)#ncol tells you how many columns will be there
plt.show()
```

The position of the label have many options. 

Here they are: 

'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'

```python
y2 = np.cos(x**2)
y3 = np.sin(x*6)
plt.figure(figsize=(8,3))
plt.plot(x,y1,ls="--", label=r'$\sin(x)$')
plt.plot(x,y, label=r'$\cos(x)$', ls="-.") # By calling plt.plot() in sequence, we can plot multiple functions on top of each other
plt.plot(x,y2, label=r'$\cos(x^2)$',ls=":")
plt.plot(x,y3, label=r'$\cos(6x)$')
plt.xlabel('x-axis', fontsize=16)
plt.ylabel('y-axis')
plt.title('Plots of Different Functions')
plt.legend(loc='lower right', fontsize=12, ncol=2) # ncol tells you the how many columns will be there
plt.show()
```

To save an image use `savefig`

```python
plt.figure(figsize=(8,3))
plt.plot(x,y1,ls="--", label=r'$\sin(x)$')
plt.plot(x,y, label=r'$\cos(x)$', ls="-.")
plt.plot(x,y2, label=r'$\cos(x^2)$',ls=":")
plt.plot(x,y3, label=r'$\cos(6x)$')
plt.xlabel('x-axis', fontsize=16)
plt.ylabel('y-axis')
plt.title('Plots of Different Functions')
plt.legend(loc='lower right', fontsize=12, ncol=2)#ncol tells you the how many columns will be there
plt.show()
plt.savefig("plot.pdf")# this save the plot in the folder where you are running your notebook
#with the name plot in pdf format.
```

### Scatter Plots

Scatter Plot means we are only plotting data points.

```python
x = np.linspace(0,15,30)
y = np.sin(x) + 0.1*np.random.randn(len(x))
x1 = np.linspace(0, 15, 100)
y1 = np.sin(x1)

plt.figure(figsize=(8,3))
plt.scatter(x,y,label='Data')
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Voltage [V]')
plt.title('Voltage in 3rd Electrode')
plt.legend(loc='upper right', fontsize=12, ncol=2)
plt.ylim(top=2)
plt.show()
```

## SubFigures

```python
# This plotting method allows you to create multiple diagrams in the same window. 
fig, axes = plt.subplots(1, 2, figsize=(16,8))#1*2= 2 subplots will be created
#-----------------------------------------------------------------------------
ax = axes[0]
ax.plot(x, y, 'o--', color='r', lw=0.4, ms=3)
ax.set_xlabel('x-axis for 1st plot')
#-----------------------------------------------------------------------------
ax = axes[1]
ax.hist(res, bins=30, density=True, histtype='step', label='C1')
#-----------------------------------------------------------------------------
fig.tight_layout()#reduce the white region from boundary
fig.suptitle('Title of All Plots', y=1.03, fontsize=25)
plt.show()
```

### Error bars

Sometimes in data we have some error. So, to correctly specify the value we add the error bars.

```python
fig = plt.figure()
x = np.arange(10)
y = 2.5 * np.sin(x / 20 * np.pi)
yerr = np.linspace(0.05, 0.2, 10)

plt.errorbar(x, y + 3, yerr=yerr, label='both limits (default)')

plt.errorbar(x, y + 2, yerr=yerr, uplims=True, label='uplims=True')

plt.errorbar(x, y + 1, yerr=yerr, uplims=True, lolims=True,
             label='uplims=True, lolims=True')

upperlimits = [True, False] * 5
lowerlimits = [False, True] * 5
plt.errorbar(x, y, yerr=yerr, uplims=upperlimits, lolims=lowerlimits,
             label='subsets of uplims and lolims')

plt.legend(loc='lower right',fontsize=12)
plt.show()
```

## Application in Astronomy

Matplotlib is often used in astronomy for data visualization.

Let's see a few examples

```python
from astropy.io import fits
from astropy.wcs import WCS
```

```python
hdul = fits.open('s0002.fits')
data = hdul[0].data
h1 = hdul[0].header
obj_name = h1.get('OBJECT', 'Unknown')

flux = data[0]
w = WCS(h1, naxis=1, relax=False, fix=False)
wvl = w.wcs_pix2world(np.arange(len(flux)), 0)[0]

plt.plot(wvl, flux)

plt.axvline(x=5895.6, color='red', linestyle='--', label='Na Absorption')
plt.axvline(x=4364.436, color='black', linestyle='--', label=r'$O_{III}$ Emissive')
plt.legend()
plt.xlabel('Wavelength (Angstrom)')
plt.ylabel('Normalized flux')
plt.show()
```

This at once give us so much information. We can see the dips, they are absorption lines. See, just by looking into the data we cannot say much. But now just by looking into this we can say so much about it.

This is not all, we can do more. We can plot H-R diagrams and visualize the data.

```python
from astroquery.gaia import Gaia
from astropy.table import Table
```

```python
tables = Gaia.load_tables(only_names = True)
for table in tables:
    if 'gaiadr3' in table.get_qualified_name():
        print(table.get_qualified_name())
```

```python
query = """SELECT source_id, ra, dec, parallax, phot_g_mean_mag, phot_bp_mean_mag, phot_rp_mean_mag FROM gaiadr3.gaia_source WHERE parallax > 50"""
job = Gaia.launch_job_async(query)
gtable = job.get_results()
```

```python
print(gtable.info)
```

```python
bp_rp = gtable['phot_bp_mean_mag'] - gtable['phot_rp_mean_mag']
fig, ax = plt.subplots(1,1,figsize=(7,7))
ax.hexbin(bp_rp, gtable['phot_g_mean_mag'],bins='log', mincnt=1)
ax.set_xlabel(r"$G_{BP} - G_{RP}$")
ax.set_ylabel(r"G")
ax.set_title("Hertzsprung Russell Diagram Before Filtering from Gaia3")
ax.invert_yaxis()
plt.show()
```

```python
abs_gmag = 5*np.log10(gtable['parallax']) - 10 + gtable['phot_g_mean_mag']
fig, ax = plt.subplots(1,1,figsize=(7,7))
ax.hexbin(bp_rp, abs_gmag,bins='log', mincnt=1)
ax.set_xlabel(r"$G_{BP} - G_{RP}$")
ax.set_ylabel(r"G")
ax.set_title("Hertzsprung Russell Diagram After Filtering from Gaia3")
ax.invert_yaxis()
plt.show()
```

## Exercise 1

Use Numpy to generate 10000 random values following a Gaussian/Normal distribution, and make a histogram. Try changing    the number of bins to properly see the Gaussian. Try overplotting a Gaussian function on top of it using a colored line, and adjust the normalization so that the histogram and the line are aligned

```python
# solution: Exercise 1
x = np.random.randn(10000) # Generate 10000 random values from a normal distribution
plt.hist(x, bins=50, density=True, label='Data') # Plot a histogram with 50 bins and (density=True) 
plt.xlabel('Value') # Label the x-axis

# Overplot a Gaussian function
mu, sigma = np.mean(x), np.std(x) # Calculate the mean and standard deviation of the data
x_fit = np.linspace(-4, 4, 100) # Generate 100 values between -4 and 4 (net search space)
y_fit = 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-(x_fit - mu)**2/(2*sigma**2)) # Calculate the Gaussian function (net search space)
plt.plot(x_fit, y_fit, color='red', label='Gaussian fit')  # Plot the Gaussian function
plt.legend() # Add a legend to the plot
plt.show() # Show the plot
```

## Exercise 2

The [central limit theorem](http://en.wikipedia.org/wiki/Central_limit_theorem) states that the arithmetic mean of a large 
number of independent random samples (from any distribution) will approach a normal distribution. You can easily test this 
with Numpy and Matplotlib:

1. Create an empty array ``total`` with 10000 values (set to 0)
2. Generate 10000 random values uniformly between 0 and 1
3. Add these values to the ``total`` array
4. Repeat steps 2 and 3 10 times
5. Divide ``total`` by 10 to get the mean of the values you added
5. Make a histogram of the values in ``total``

You can also see how the histogram of ``total`` values changes at each step, if you want to see the evolution!

```python
# solution: Exercise 2

# 1.Create an empty array with 10000 values (set to 0)
total = np.zeros(10000)

# 2.Generate 10000 random values uniformly between 0 and 1
for i in range(10):
    random_values = np.random.rand(10000)
    # 3.Add these values to the total array
    total += random_values
    
    # 4. repeat steps 2 and 3 10 times in for loop
    
# 5.Divide total by 10 to get the mean of the values you added
total /= 10

# 6.Make a histogram of the values in total
plt.hist(total, bins=50, label='Data', color="red") # Plot a histogram with 50 bins
plt.xlabel('Value') # Label the x-axis
plt.show() # Show the plot

```
